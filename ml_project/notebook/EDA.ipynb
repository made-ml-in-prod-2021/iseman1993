```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DATA_PATH = '../data/raw/heart.csv'
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
CAT_COLS = [
    "sex",
    "cp",
    "fbs",
    "restecg",
    "exang",
    "slope",
    "ca",
    "thal",
]

NUM_COLS = [
    "age",
    "trestbps",
    "chol",
    "thalach",
    "oldpeak",
]
```

```python
sns.pairplot(df[NUM_COLS + ["target"]], hue="target")
```

```python
plt.figure(figsize=(16, 16))
sns.heatmap(df.corr().abs(), annot=True, linewidths=0.5)
```

```python
for col in CAT_COLS:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=df[col], hue=df["target"])
    plt.legend(loc="best", title="target")
    plt.grid()
    plt.show()
```
